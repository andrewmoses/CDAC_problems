```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

iris = datasets.load_iris()
# iris = datasets.load_iris(return_X_y=True)
iris.feature_names
```

```python
#
X = iris.data
y = iris.target
print("target", y)
print("independant variables: ", X)
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
kmeans_oly2 = KMeans(n_clusters=3, random_state=0).fit(X[:,:2])
print(kmeans)
print(kmeans_oly2)
print(kmeans.labels_)
print(kmeans_oly2.labels_)
print(kmeans.cluster_centers_)
```

```python
kmeans_y = kmeans.labels_
kmeans_y_oly2 = kmeans_oly2.labels_
print(kmeans.labels_)
# all 1's should become 0
# all 0's shoudl become 2
# all 2's shoudl become 1
mod_kmeans_y = []
for each in kmeans_y:
    if each==1:
        mod_kmeans_y.append(0)
    elif each==0:
        mod_kmeans_y.append(2)
    elif each==2:
        mod_kmeans_y.append(1)
        
mod_kmeans_y_oly2 = []
for each in kmeans_y_oly2:
    if each==1:
        mod_kmeans_y_oly2.append(0)
    elif each==0:
        mod_kmeans_y_oly2.append(1)
    elif each==2:
        mod_kmeans_y_oly2.append(2)
# plt.scatter(X[:,0], X[:,1], c=kmeans_y)
# plt.scatter(X[:,0], X[:,1], c=mod_kmeans_y)
print(mod_kmeans_y_oly2)
print(mod_kmeans_y)
plt.subplot(1, 3, 1)
plt.scatter(X[:,0], X[:,1], c=y)
plt.subplot(1, 3, 2)
plt.scatter(X[:,0], X[:,1], c=mod_kmeans_y_oly2)
plt.subplot(1, 3, 3)
plt.scatter(X[:,0], X[:,1], c=mod_kmeans_y)
```
