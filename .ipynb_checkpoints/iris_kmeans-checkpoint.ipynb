```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

iris = datasets.load_iris()
iris
```

```python
#
X = iris.data
y = iris.target
print("target", y)
print("independant variables: ", X)
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
# kmeans = KMeans(n_clusters=4, random_state=0).fit(X[:,:2])
print(kmeans)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
# target and predicators
# [: 2] takes only two columns
X = iris.data[:,:2]
y = iris.target
# color is from y values.. e.g. 0 : dark voilent, 1: green, 2: yellow
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
```

```python
kmeans_y = kmeans.labels_
print(kmeans.labels_)
# all 1's should become 0
# all 0's shoudl become 2
# all 2's shoudl become 1
mod_kmeans_y = []
for each in kmeans_y:
    if each==1:
        mod_kmeans_y.append(0)
    elif each==0:
        mod_kmeans_y.append(2)
    elif each==2:
        mod_kmeans_y.append(1)
plt.scatter(X[:,0], X[:,1], c=kmeans_y)
plt.scatter(X[:,0], X[:,1], c=mod_kmeans_y)

plt.subplot(2, 1, 1)
```
