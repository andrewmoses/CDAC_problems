```python
# tic tac toe
board = [['-','-','-'],
         ['-','-','-'],
         ['-','-','-']]
#check if someone won or not
def haswon(board):
    """
    given a board, check for row, column and diagonal
    """
    # row
    won = False
    ele_count = {'o':0, 'x':0}
    for row in board:
        for item in row:
            if not item == '-':
                if item == 'x':
                    ele_count['x'] = ele_count['x'] + 1
                else:
                    ele_count['o'] = ele_count['o'] + 1
        #check if x or o won in this row
        if ele_count['o']==3:
            print('o has won!')
            return True
        elif ele_count['x']==3:
            print('x has won!')
            return True
        else:
            # initialize ele_count again
            ele_count = {'o':0, 'x':0}
    
    #column
    ele_count = {'o':0, 'x':0}
    for x in range(3):
        for y in range(3):
            item = board[y][x]
            if not item=='-':
                if item == 'x':
                    ele_count['x'] = ele_count['x'] + 1
                else:
                    ele_count['o'] = ele_count['o'] + 1
        # check if x or o won in this column
        if ele_count['o']==3:
            print('o has won!')
            return True
        elif ele_count['x']==3:
            print('x has won!')
            return True
        else:
            # initialize ele_count again
            ele_count = {'o':0, 'x':0}
    # diagonal
    ele_count = {'o':0, 'x':0}
    d_l = 0
    for d in range(3):
        item = board[d_l][d_l]
        if not item=='-':
            if item == 'x':
                ele_count['x'] = ele_count['x'] + 1
            else:
                ele_count['o'] = ele_count['o'] + 1
        d_l = d_l + 1
    # check if x or o won in this column
    if ele_count['o']==3:
        print('o has won!')
        return True
    elif ele_count['x']==3:
        print('x has won!')
        return True
    else:
        # initialize ele_count again
        ele_count = {'o':0, 'x':0}
    # diagonal right to left
    dx = 0
    dy = 2
    for d in range(3):
        item = board[dx][dy]
        if not item=='-':
            if item == 'x':
                ele_count['x'] = ele_count['x'] + 1
            else:
                ele_count['o'] = ele_count['o'] + 1
        dx = dx + 1
        dy = dy - 1
    # check if x or o won in this column
    if ele_count['o']==3:
        print('o has won!')
        return True
    elif ele_count['x']==3:
        print('x has won!')
        return True
    else:
        # initialize ele_count again
        ele_count = {'o':0, 'x':0}
        
    return False

board = [['x','o','o'],
         ['o','o','x'],
         ['o','x','x']]
print(haswon(board))

# def pick_empty_space(board):
import random
def pick_empty_space(board):
    #check if board has empty places
    flag = 0
    for row in board:
        for item in row:
            if item=='-':
                flag = 1
                break
    if flag==0:
        print("Board is filled! Game over")
        return False
    while True:
        x = random.randint(0,2)
        y = random.randint(0,2)
        # check if its empty space
        if board[x][y] == '-':
            return x,y
# pick_empty_space(board)
def fill_board(ip_val, board, x, y):
    """
    given x or o and board empty position
    fill it up
    """
    board[x][y] = ip_val

board = [['x','x','x'],
         ['-','-','-'],
         ['x','x','o']]
# fill_board('x',board, 1,0)
# print(board)
def displayboard(board):
    print("\n")
    print("*"*5)
    for row in board:
        for item in row:
            print(item, end=" ")
        print("")
    print("*"*5)
    print("\n")
# displayboard(board)
```

```python
turn = 'o'
board = [['-','-','-'],
         ['-','-','-'],
         ['-','-','-']]
while True:
    # pick empty space
    if not pick_empty_space(board):
        break
    print(turn+" Turn!!: ")
    x,y = pick_empty_space(board)
    fill_board(turn, board, x, y)
    displayboard(board)
    if haswon(board):
        break
    if turn == 'o':
        turn = 'x'
    else:
        turn = 'o'

# while True:
#     if turn == 'o':
#         turn = 'x'
#         # pick empty space
#         if not pick_empty_space(board):
#             break
#         print(turn+" Turn!!: ")
# #         x,y = pick_empty_space(board)
#         x = int(input('position of x: '))
#         y = int(input('position of y: '))
#         fill_board(turn, board, x, y)
#         displayboard(board)
#         if haswon(board):
#             break
        
#     else:
#         turn = 'o'
#         # pick empty space
#         if not pick_empty_space(board):
#             break
#         print(turn+" Turn!!: ")
#         x,y = pick_empty_space(board)
#         fill_board(turn, board, x, y)
#         displayboard(board)
#         if haswon(board):
#             break
    
    
```
