```python
import pandas as pd
bw_df = pd.read_csv('balance-scale.data', header=None)
bw_df
```

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(bw_df.values[:,1:5], bw_df.values[:, 0], test_size=0.30)
# y_test
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
clf_en = tree.DecisionTreeClassifier(criterion='entropy')
clf_en.fit(x_train,y_train)
predicted_y = clf_en.predict(x_test)
print(accuracy_score(y_test, predicted_y))

clf_gi = tree.DecisionTreeClassifier(criterion='gini', min_samples_split=2)
clf_gi.fit(x_train,y_train)
predicted_y = clf_gi.predict(x_test)
print(accuracy_score(y_test, predicted_y))
```
