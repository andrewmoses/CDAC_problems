```python
import pandas as pd
df = pd.read_csv("mnist_kaggle.csv")
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def visualize_image(images_5max):
    reshaped_images = []
    for i in range(len(images_5max)):
        reshaped_images=images_5max[i,:].reshape(28,28)
        plt.imshow(reshaped_images, cmap ='gist_gray')
        plt.show()

visualize_image(df.values[:5,1:])
```

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(df.values[:,1:], df.values[:, 0], test_size=0.30)

from sklearn.neural_network import MLPClassifier
clf = MLPClassifier().fit(x_train, y_train)

from sklearn.metrics import accuracy_score
predicted_y = clf.predict(x_test)
print(accuracy_score(y_test, predicted_y))
```

```python
# get the indexes of wrongly computed labels
wrongly_done = []
for each in range(len(predicted_y)):
    if not predicted_y[each] == y_test[each]:
        wrongly_done.append(each)
print("total number of images incorrectly predicted: ", len(wrongly_done))
```

```python
# visualize the incorrectly predicted
for w_i in wrongly_done[:5]:
    print("Original label: ", y_test[w_i])
    print("Predicted label: ", predicted_y[w_i])
    visualize_image(x_test[w_i:w_i+1,:])
    print("\n\n")
```
